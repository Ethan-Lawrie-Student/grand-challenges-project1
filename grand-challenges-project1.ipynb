```python
import pandas as pd
```

```python
df = pd.read_csv("TMDB_movie_dataset_v11.csv")
```

```python
print(df)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d')
print(df['release_date'])
```

```python
filtered_df = df.loc[(df['release_date'] >= '2024-07-29')
                     & (df['release_date'] < '2024-07-30')]
# Display
filtered_df
```

```python
filtered_df = filtered_df.drop(filtered_df[filtered_df['status'] == 'In Production'].index)
filtered_df
```

```python
filtered_df = filtered_df.drop(filtered_df[filtered_df['status'] != 'Released'].index)
filtered_df
```
